```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset into a DataFrame
# Replace 'your_dataset.csv' with the actual path to your filew-bipinsebastian/projects/project 4/global_youtube_statistics.csv')
channel_data = pd.read_csv(r'C:/Users/Gymnasiet/floeder for python/data-hw-bipinsebastian/projects/project 4/global_youtube_statistics.csv', encoding='ISO-8859-1')

# Preview the DataFrame to ensure the data is correctly loaded
print(channel_data.head())
```

 Show in a graph the 10 channels with the most subscribers

```python


# Assuming 'channel_data' is the DataFrame with 'channel' and 'subscribers' columns
top_10_subscribers = channel_data.nlargest(10, 'subscribers')

plt.figure(figsize=(10,6))
plt.bar(top_10_subscribers['Youtuber'], top_10_subscribers['subscribers'], color='skyblue')
plt.title('Top 10 Channels with Most Subscribers')
plt.xlabel('Channels')
plt.ylabel('Subscribers (in billions)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
 Show in a graph the 10 channels with the least subscribers.
```

```python
bottom_10_subscribers = channel_data.nsmallest(10, 'subscribers')

plt.figure(figsize=(10,6))
plt.bar(bottom_10_subscribers['Youtuber'], bottom_10_subscribers['subscribers'], color='salmon')
plt.title('Top 10 Channels with Least Subscribers')
plt.xlabel('Channels')
plt.ylabel('Subscribers (in millions)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

 Show the 10 Most-Viewed YouTube Channels

```python

top_10_views = channel_data.nlargest(10, 'video views')

plt.figure(figsize=(10,6))
plt.bar(top_10_views['Youtuber'], top_10_views['video views'], color='lightgreen')
plt.title('Top 10 Most-Viewed YouTube Channels')
plt.xlabel('Channels')
plt.ylabel('Views (in billions)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import seaborn as sns

category_counts = channel_data['category'].value_counts()

plt.figure(figsize=(12,6))
sns.barplot(x=category_counts.index, y=category_counts.values, palette='viridis')
plt.title('Number of Channels in Each Category')
plt.xlabel('Category')
plt.ylabel('Number of Channels')
plt.xticks(rotation=45, ha='right')
plt.show()
```

 Show a Line Graph of the 10 Channels with the Most Uploads and Subscribers

```python
top_10_uploads = channel_data.nlargest(10, 'uploads')

plt.figure(figsize=(10,6))
plt.plot(top_10_uploads['Youtuber'], top_10_uploads['uploads'], marker='o', label='Uploads')
plt.plot(top_10_uploads['Youtuber'], top_10_uploads['subscribers'], marker='x', label='Subscribers', color='orange')
plt.title('Uploads vs Subscribers for Top 10 Channels')
plt.xlabel('Channels')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.show()



```

Show a Graph of the Number of Channels per Country

```python
country_counts = channel_data['Country'].value_counts()

plt.figure(figsize=(12,6))
sns.barplot(x=country_counts.index, y=country_counts.values, palette='magma')
plt.title('Number of Channels by Country')
plt.xlabel('Country')
plt.ylabel('Number of Channels')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Show a Graph of the Number of Channels Created per Year

```python
channel_data['creation_year'] = pd.DatetimeIndex(channel_data['created_date']).year
yearly_counts = channel_data['creation_year'].value_counts().sort_index()

plt.figure(figsize=(12,6))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o')
plt.title('Number of Channels Created Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Channels')
plt.xticks(rotation=45)
plt.show()
```
